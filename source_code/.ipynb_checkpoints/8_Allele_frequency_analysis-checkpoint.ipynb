# Analyze allele frequencies of variants in the database

```python
import pandas as pd
import requests, sys
import time
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(
    '/Users/ksenia/Documents/MODY_genes/whole_pipeline_311022/MD_variants/intermediate/MD_variants_forVCF_level1.csv')
data
#MD_variants_forVCF_level2.csv
```

```python
server = "https://rest.ensembl.org"
no_info = []
json_list = []
for var in data['accession']:
    if var.startswith('rs'):
        ext = "/ga4gh/variants/1:" + var + "?"
        r = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
 
        if not r.ok:
            print(var)
            no_info.append('bad' + var)
            continue
            
        time.sleep(1)
        json_list.append(r.json())
        print('ok' + var)
```

```python
AF = []
AF_EUR = []
for variant in json_list:
    if 'info' in variant.keys():
        AF.append(variant['info']['AF'])
        AF_EUR.append(variant['info']['EUR_AF'])
```

for variant in json_list:

```python
AF_float = []
for item in AF:
    for value in item:
        if ',' in value:
            print(value)
            for i in value.split(','):
                AF_float.append(float(i))
        else: AF_float.append(float(value))
AF_float[0]
```

```python
len(AF_EUR)
```

```python
AF_EUR_float = []
for item in AF_EUR:
    for value in item:
        if ',' in value:
            print(value)
            for i in value.split(','):
                AF_EUR_float.append(float(i))
        else: AF_EUR_float.append(float(value))
AF_EUR_float[0]
```

```python
plt.plot(sorted(AF_float, key=float),'g*', sorted(AF_EUR_float, key=float), 'ro')
plt.show()
```

```python
rare_AF = []
low_frequency_AF = []
common_AF = []
for freq in AF_float:
    if freq < 0.01:
        rare_AF.append(freq)
    if 0.01 < freq < 0.05:
        low_frequency_AF.append(freq)
    if  freq > 0.05:
        common_AF.append(freq)
```

```python
plot_data = {'rare': len(rare_AF), 'low-frequency': len(low_frequency_AF), 'common_AF': len(common_AF)}
names = list(plot_data.keys())
values = list(plot_data.values())
plt.bar(names, values)
```

```python
labels = 'rare', 'low-frequency', 'common'
sizes = [len(rare_AF), len(low_frequency_AF), len(common_AF)]
colors = ['#FF9347', '#48B54B','#5F49B3']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90, colors=colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
data2 = pd.read_csv(
    '/Users/ksenia/Documents/MODY_genes/whole_pipeline_311022/MD_variants/intermediate/MD_variants_forVCF_level2.csv')
server = "https://rest.ensembl.org"
json_list2 = []
count = 0
for var in data2['accession']:
    if var.startswith('rs'):
        ext = "/ga4gh/variants/1:" + var + "?"
        r2 = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
 
        if not r2.ok:
            print('bad' + var)
            continue
            
        time.sleep(1)
        json_list2.append(r2.json())
        print('ok' + var)
        count = count + 1
        print(count)

AF2 = []
AF_EUR2 = []
for variant in json_list2:
    if 'info' in variant.keys():
        AF2.append(variant['info']['AF'])
        AF_EUR2.append(variant['info']['EUR_AF'])
AF_float2 = []
for item in AF2:
    for value in item:
        if ',' in value:
            print(value)
            for i in value.split(','):
                AF_float2.append(float(i))
        else: AF_float2.append(float(value))

```

```python
rare_AF2 = []
low_frequency_AF2 = []
common_AF2 = []
for freq in AF_float2:
    if freq < 0.01:
        rare_AF2.append(freq)
    if 0.01 < freq < 0.05:
        low_frequency_AF2.append(freq)
    if  freq > 0.05:
        common_AF2.append(freq)
        
labels2 = 'rare', 'low-frequency', 'common'
sizes2 = [len(rare_AF2), len(low_frequency_AF2), len(common_AF2)]
colors2 = ['#FF9347', '#48B54B','#5F49B3']
fig12, ax12 = plt.subplots()
ax12.pie(sizes2, labels=labels2, autopct='%1.1f%%',
        shadow=False, startangle=90, colors=colors2)
ax12.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
